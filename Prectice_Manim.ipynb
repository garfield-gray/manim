```python
%%manim -qh -v WARNING Basic

class Basic(Scene):
  def construct(self):
    polys = VGroup(
        *[RegularPolygon(3, radius = 1, fill_opacity = 0.8) for j in range(6)]
    ).arrange(RIGHT)
    self.play(DrawBorderThenFill(polys), run_time=2)
    self.play(
        # Rotate(polys[0], PI, rate_func=lambda t: t),
        # Rotate(polys[1], PI, rate_func=smooth),
        # Rotate(polys[2], PI, rate_func=lambda t: np.sin(t*np.pi)),
        Rotate(polys[3], PI, rate_func=there_and_back),
        # Rotate(polys[4], PI, rate_func=lambda t: np.sin(t*np.pi)),
        Rotate(polys[5], PI, rate_func=lambda t: t**2),
        run_time=2
    )
    self.wait(2)

```

```python
%%manim -qh -v WARNING Lagging

class Lagging(Scene):
  def construct(self):
    squares = VGroup(
        *[Square(color = Color(hue=j/12, saturation=1,luminance=0.5),
        fill_opacity=0.75)for j in range(12)]).arrange_in_grid(3,4)
    self.play(AnimationGroup(*[FadeIn(s) for s in squares], lag_ratio=0.1, rate_func=lambda t: np.sqrt(t)))

    self.wait(2)

```

```python
# .animate problems (only cares about start and finish) and target and restore
%%manim -qh -v WARNING Animate

class Animate(Scene):
  def construct(self):
    s1, s2 = Square(), Square()
    self.add(s1,s2)
    self.play(s1.animate.rotate(PI), Rotate(s2, PI), run_time=3)

    # s = VGroup(s1, s2).arrange(UP, buff = 0.5)
    s1.save_state()
    s1.generate_target()
    s1.target.set_fill(color=GREEN, opacity=0.9)
    s1.target.shift(2*RIGHT).scale(0.4)
    self.play(MoveToTarget(s1))
    self.play(Rotate(s1, 0.5*PI), run_time=2)
    s2.save_state()
    self.play(Restore(s1), s2.animate.rotate(0.1*PI).shift(3*LEFT))
    self.play(Restore(s2))


    self.wait(2)

```
