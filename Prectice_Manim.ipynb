```python
%%manim -qh -v WARNING Basic

class Basic(Scene):
  def construct(self):
    polys = VGroup(
        *[RegularPolygon(3, radius = 1, fill_opacity = 0.8) for j in range(6)]
    ).arrange(RIGHT)
    self.play(DrawBorderThenFill(polys), run_time=2)
    self.play(
        # Rotate(polys[0], PI, rate_func=lambda t: t),
        # Rotate(polys[1], PI, rate_func=smooth),
        # Rotate(polys[2], PI, rate_func=lambda t: np.sin(t*np.pi)),
        Rotate(polys[3], PI, rate_func=there_and_back),
        # Rotate(polys[4], PI, rate_func=lambda t: np.sin(t*np.pi)),
        Rotate(polys[5], PI, rate_func=lambda t: t**2),
        run_time=2
    )
    self.wait(2)

```

```python
%%manim -qh -v WARNING Lagging

class Lagging(Scene):
  def construct(self):
    squares = VGroup(
        *[Square(color = Color(hue=j/12, saturation=1,luminance=0.5),
        fill_opacity=0.75)for j in range(12)]).arrange_in_grid(3,4)
    self.play(AnimationGroup(*[FadeIn(s) for s in squares], lag_ratio=0.1, rate_func=lambda t: np.sqrt(t)))

    self.wait(2)

```

```python
# .animate problems (only cares about start and finish) and target and restore
%%manim -qh -v WARNING Animate

class Animate(Scene):
  def construct(self):
    s1, s2 = Square(), Square()
    self.add(s1,s2)
    self.play(s1.animate.rotate(PI), Rotate(s2, PI), run_time=3)

    # s = VGroup(s1, s2).arrange(UP, buff = 0.5)
    s1.save_state()
    s1.generate_target()
    s1.target.set_fill(color=GREEN, opacity=0.9)
    s1.target.shift(2*RIGHT).scale(0.4)
    self.play(MoveToTarget(s1))
    self.play(Rotate(s1, 0.5*PI), run_time=2)
    s2.save_state()
    self.play(Restore(s1), s2.animate.rotate(0.1*PI).shift(3*LEFT))
    self.play(Restore(s2))


    self.wait(2)

```

```python
# UpdateFromAlphaFunc
%%manim -qh -v WARNING Custom

class Custom(Scene):
  def construct(self):
    def SpiralOut(mobject, t):
      radius = 3*t
      angle = 4*t*PI

      mobject.move_to(radius*(np.cos(angle)*RIGHT + np.sin(angle)*UP))
      mobject.set_color(Color(hue=t, saturation=1, luminance = 0.5))
      mobject.set_opacity(1-0.5*t)

    # d = Dot(color=WHITE)
    d = Square(side_length=0.5)
    self.add(d)
    self.play(UpdateFromAlphaFunc(d, SpiralOut, run_time=10))







```

```python
# Anatomy of Animation

%%manim -qh -v WARNING MyAnim



class Disperse(Animation):
  def __init__(self, mobject, dot_radius=0.05, dot_number=100, **kwargs):            #override of __init__
    super().__init__(mobject, **kwargs)            # calling the initialization with thaparent

    self.dot_radius = dot_radius
    self.dot_number = dot_number


  def begin(self):
    dots = VGroup(
        *[Dot(radius=self.dot_radius).move_to(self.mobject.point_from_proportion(p))
        for p in np.linspace(0, 1, self.dot_number)]
    )

    for dot in dots:
      dot.initial_position = dot.get_center()
      dot.shift_vector = 2*(dot.get_center()-self.mobject.get_center())

    dots.set_opacity(0)
    self.mobject.add(dots)
    self.dots = dots

    super().begin()

  def clean_up_from_scene(self, scene: "Scene") -> None:
    super().clean_up_from_scene(scene)
    scene.remove(self.dots)



  def interpolate_mobject(self, alpha):
    alpha = self.rate_func(alpha)
    if alpha <= 0.5:
      self.mobject.set_opacity(1-2*alpha, family=False)
      self.dots.set_opacity(2*alpha)
    else:
      self.mobject.set_opacity(0)
      self.dots.set_opacity(2*(1-alpha))
      for dot in self.dots:
        dot.move_to(dot.initial_position + 2*(alpha - 0.5)*dot.shift_vector)





class MyAnim(Scene):
  def construct(self):
    st = Star(color=YELLOW, fill_opacity=1).scale(3)

    self.add(st)
    self.wait()
    self.play(Disperse(st, dot_number=200, run_time = 5))







```
