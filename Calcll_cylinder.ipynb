```python
from manim import *
from manim.opengl import *
```

```python
%%manim -qh -v WARNING cylinder




class cylinder(ThreeDScene):
    def func(self, u, v):
      return np.array([np.cos(u) , np.sin(u), v])
    def construct(self):
        axes = ThreeDAxes()

        x_label = axes.get_x_axis_label('x')
        y_label = axes.get_y_axis_label('y').shift(UP * 1.8)
        z_label = axes.get_z_axis_label('z')

        self.play(FadeIn(axes), FadeIn(x_label), FadeIn(y_label), FadeIn(z_label))

        self.wait(0.5)

        # animate the move of the camera to properly see the axes
        self.move_camera(phi=75 * DEGREES, theta=30 * DEGREES, zoom=1, run_time=1.5)

        # generating the surface
        surface = Surface(
            lambda u, v: axes.c2p(*self.func(u, v)),
            u_range=[0, 2*PI],
            v_range=[-1, 1],
            resolution=8,
        )
        self.play(FadeIn(surface))

        #adding Title
        text01 = MathTex("R^2 = x^2 + y^2")
        text01.to_corner(DL)
        #self.play(FadeIn(surfacecylinder))
        self.add_fixed_in_frame_mobjects(text01)
        self.wait(1)

        #removing title
        self.remove_fixed_in_frame_mobjects(text01)
        self.wait(0.1)

        #diffrent angles




        #x-y
        self.move_camera(phi=0 * DEGREES, theta=0 * DEGREES, zoom=1, run_time=1.5)

        text02 = MathTex("for\ same\ amount\ of\ z \ : X^2+y^2=R^2").scale(0.75)
        text02.to_corner(DL)
        self.add_fixed_in_frame_mobjects(text02)
        self.wait(3)
```
