```python
%%manim -qh -v WARNING Eigen

class Eigen(Scene):
  def construct(self):
    Question = Tex("Find the Eigen Values, Eigen Vectors and jordan form of")

    M = Matrix([[1 ,0 ,0],[0 ,2 ,2 ],[0 , 1 , 3]])
    M1 = M.copy()
    M2 = M.copy()
    M3 = M.copy()
   
    self.play(Create(Question), run_time = 5)
    self.play(Question.animate.shift(3*UP))
    self.play(Create(M), run_time = 2)
    self.wait(0.5)
    self.play(FadeOut(Question, scale = 0.3),M.animate.shift(2*(UP+LEFT)), run_time=1.5)
    formula = MathTex(r"det ( A - \lambda I ) = 0").shift(2*(UP+RIGHT))
    self.play(Create(formula), run_time = 2)
    Mlambda = MathTex(r"det \begin{bmatrix} 1-\lambda & 0 & 0 \\ 0 & 2-\lambda & 2 \\0 & 1 & 3-\lambda \end{bmatrix} = 0")
    PreMlambda = VGroup(M, formula)
    self.play(TransformMatchingShapes(PreMlambda, Mlambda), run_time = 5)
    plainlambda = MathTex(r"(1-\lambda)[(2-\lambda)(3-\lambda)-2]=0")
    plainlambda2 = MathTex(r"(1-\lambda)",r"[\lambda^2-5\lambda+4]=0")
    plainlambda3 = MathTex(r"(1-\lambda)",r"(4-\lambda)",r"(1-\lambda)=0")
    
    v1 = Vector([0,-1,0]).next_to(plainlambda3[0], DOWN).shift(3*UP+4*LEFT)
    v2 = v1.copy().next_to(plainlambda3[1], DOWN).shift(3*UP+4*LEFT)
    v3 = v2.copy().next_to(plainlambda3[2], DOWN).shift(3*UP+4*LEFT)
    
    Lambda1 = MathTex(r"\lambda_1 = 1").align_to(v1, DOWN+LEFT).shift(0.5*DOWN)
    Lambda2 = MathTex(r"\lambda_2 = 4").align_to(v2, DOWN+LEFT).shift(0.5*DOWN)
    Lambda3 = MathTex(r"\lambda_3 = 1").align_to(v3, DOWN+LEFT).shift(0.5*DOWN)

    self.play(TransformMatchingShapes(Mlambda, plainlambda), run_time = 5)
    self.play(plainlambda.animate.shift(2.5*LEFT))
    self.play(TransformMatchingShapes(plainlambda, plainlambda2), run_time = 2)
    self.play(TransformMatchingTex(plainlambda2, plainlambda3), run_time = 1)
    self.play(plainlambda3.animate.shift(3*UP+4*LEFT))

    self.play(Create(v1),Create(v2),Create(v3), run_time=2)
    self.play(Create(Lambda1), run_time = 2)
    self.play(Create(Lambda2), run_time = 2)
    self.play(Create(Lambda3), run_time = 2)

    eigens = VGroup(plainlambda3, v1,v2,v3,Lambda1,Lambda2,Lambda3)

    Lambda1and3 = MathTex(r"\lambda_{1\&3} = 1").align_to(Lambda1, LEFT+DOWN)
    self.play(ReplacementTransform(eigens, Lambda1and3), run_time = 2)

    lambdaformula = MathTex(r"\; , \; AX = \lambda X").next_to(Lambda1and3, RIGHT)
    self.play(Write(lambdaformula), run_time=2)
    stuff = VGroup(Lambda1and3, lambdaformula)
    f = MathTex(r"AX = X").next_to(stuff, RIGHT)
    self.play(ReplacementTransform(stuff, f), run_time=3)


    X = Matrix([['x_1'],['x_2'],['x_3']]).next_to(M3, RIGHT)
    eq = MathTex(r"=").next_to(X, RIGHT)
    X1 = X.copy().next_to(eq, RIGHT)
    ei = VGroup(M3, X,eq, X1).shift(1.5*LEFT)

    self.play(ReplacementTransform(f, ei), run_time=3)

    
    AX = Matrix([["x_1"],["2x_2+2x_3"],["x_2+3x_3"]]).align_to(X, RIGHT)

    eqq = MathTex(r"=").next_to(AX, RIGHT)
    X11 = X1.copy().next_to(eqq, RIGHT)
    
    eig = VGroup(AX, eqq, X11).shift(1.5*LEFT)

    self.play(ReplacementTransform(ei, eig), run_time=3)


    self.play(eig.animate.shift(2*UP))
    self.wait(3)
```
