```python
from manim import *
```

```python
%%manim -qh -v WARNING Eigen

class Eigen(Scene):
  def construct(self):
    WF = MathTex(r"\begin{bmatrix} x_{11} & x_{12} & x_{13} & \dots  & x_{1n} \\x_{21} & x_{22} & x_{23} & \dots  & x_{2n} \\\vdots & \vdots & \vdots & \ddots & \vdots \\x_{d1} & x_{d2} & x_{d3} & \dots  & x_{dn}\end{bmatrix}", r"\lim_{n \hookrightarrow\infty}\sum_{i=1}^{n}f(x_{i})\Delta x")
    Bi = Brace(WF[0], UP)
    ti = Bi.get_text("Integral Form")

    Br = Brace(WF[1],DOWN)
    tr = Br.get_text("$\int$ as a Riemann Sum ")

    formula = VGroup(WF,Bi,ti,Br,tr)
    self.play(Create(formula), run_time = 5)
    self.wait()
    self.play(FadeOut(formula, scale = 0.3), run_time=1.5)
    tex = Tex(r"\LaTeX", font_size=274)
    self.add(tex)
    self.wait()
```
