```python
from manim import *
```

```python
%%manim -qh -v WARNING Eigen

class Eigen(Scene):
  def construct(self):
    Question = Tex("Find the Eigen Values, Eigen Vectors and jordan form of")

    WF = MathTex(r"A = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 2 & 2 \\0 & 1 & 3 \end{bmatrix}")
   
    self.play(Create(Question), run_time = 5)
    Question.shift(3*UP)
    self.play(Create(WF), run_time = 5)
    self.wait()
    self.play(FadeOut(WF, scale = 0.3), run_time=1.5)
    self.play(FadeOut(Question, scale = 0.3), run_time=1.5)
    tex = Tex(r"\LaTeX", font_size=274)
    self.add(tex)
    self.wait(0.3)
```
