```python
from manim import *
```

```python
%%manim -qh -v WARNING SquareToCircle

config.background_color = WHITE
class SquareToCircle(Scene):
   def construct(self):
      square = Square()
      circle = Circle()
      circle.set_fill(PINK, opacity=0.9)
      square.set_fill(BLACK, opacity=0.9)
      self.play(Create(square))
      self.play(Transform(square, circle))
      self.wait()
```
