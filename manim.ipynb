```python
from manim import *
```

```python
%%manim -qm -v WARNING SquareToCircle

#config.media_width = "20"
class SquareToCircle(Scene):
   def construct(self):
      square = Square()
      circle = Circle()
      circle.set_fill(PINK, opacity=0.9)
      self.play(Create(square))
      self.play(Transform(square, circle))
      self.wait()
```
