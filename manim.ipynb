#Latex

```python
WF = Tex("\int_{a}^{b}f(x)dx = ", "\lim_{n\rightarrow\infty}\sum_{i=1}^{n}f(x_{i})\Delta ")
```

```python
%%manim -qh -v WARNING IntTo

class IntTo(Scene):
  def construct(self):
    WF = MathTex(r"\int_{a}^{b}f(x)dx = ", r"\lim_{n \hookrightarrow\infty}\sum_{i=1}^{n}f(x_{i})\Delta x")
    Bi = Brace(WF[0], UP)
    ti = Bi.get_text("Integral Form")

    Br = Brace(WF[1],DOWN)
    tr = Br.get_text("$\int$ as a Riemann Sum ")

    formula = VGroup(WF,Bi,ti,Br,tr)
    self.play(Create(formula), run_time = 5)
    self.wait()
    self.play(FadeOut(formula, scale = 0.3), run_time=1.5)
    tex = Tex(r"\LaTeX", font_size=274)
    self.add(tex)
    self.wait()
```

#Shapes

```python
%%manim -qh -v WARNING SquareToCircle

config.background_color = BLACK
class SquareToCircle(Scene):
   def construct(self):
      square = Square()
      circle = Circle()
      circle.set_fill(RED, opacity=0.3)
      square.set_fill(GREEN, opacity=0.9)
      self.play(Create(square))
      self.play(Transform(square, circle))
      self.wait()
```
